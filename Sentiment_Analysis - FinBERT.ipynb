```python
from transformers import BertTokenizer, BertForSequenceClassification
from transformers import pipeline
import pandas as pd
```

```python
df = pd.read_csv('GoogleStockNews.csv')
df.head()
```

```python
text = df['Summary'][0]
text
```

### Sentiment Analysis

Analyzing financial text sentiment is valuable as it can engage the views and opinions of managers, information intermediaries and investors. FinBERT-Sentiment is a FinBERT model fine-tuned on 10,000 manually annotated sentences from analyst reports of S&P 500 firms.

**Input:** A financial text.

**Output:** Positive, Neutral or Negative.

```python
finbert = BertForSequenceClassification.from_pretrained('yiyanghkust/finbert-tone',num_labels= 3)
tokenizer = BertTokenizer.from_pretrained('yiyanghkust/finbert-tone')
```

```python
nlp = pipeline("sentiment-analysis", model= finbert, tokenizer= tokenizer)
results = nlp(text)
results
```

```python
sentiments = []
for i in range(len(df)):
    Sentiments = nlp(df['Summary'][i])[0].get('label')
    sentiments.append(Sentiments)
sentiments
```

```python
sentiment_series = pd.Series(sentiments, index= range(len(df)))
df['Summary Sentiments'] = sentiment_series.values
```

```python
headline_sentiments = []
for i in range(len(df)):
    Sentiments = nlp(df['Article_Headline'][i])[0].get('label')
    headline_sentiments.append(Sentiments)
headline_sentiments
```

```python
headline_series = pd.Series(headline_sentiments, index= range(len(df)))
df['Headline Sentiments'] = headline_series.values
df
```

```python
correct = 0
wrong = 0
for i in range(len(df)):
    if df['Summary Sentiments'][i] == df['Headline Sentiments'][i]:
        correct += 1
    else:
        wrong += 1
```

```python
print(f'Out model can match {correct} summary sentiments as same as headline sentiments and {wrong} summary sentiments as different from headline sentiments')
```

```python
print(f'Out model can match {correct/len(df)} summary sentiments as same as headline sentiments and {wrong/len(df)} summary sentiments as different from headline sentiments')
```
